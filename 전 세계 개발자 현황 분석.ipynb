```python
# 판다스 라이브러리 탑재
import pandas as pd

# CSV 파일 읽어오기
raw_data = pd.read_csv("C:\data\survey_results_public.csv")

# 데이터 확인하기
raw_data.info()
```

## 일부 데이터 확인하기

```python
# 일부 데이터 확인하기
raw_data.head()
```

## 필요한 열만 추출하기

```python
# 필요한 열 추출
revised_data = raw_data[["Age", "Country", "LanguageHaveWorkedWith", "LearnCode"]]

# 일부 값 확인하기
revised_data.head()
```

## 특정 열 데이터 조회하기

```python
# 연령대 정보 확인하기
revised_data["Age"]
```

## 유일 데이터 확인하기

drop_duplicates() 메서드 실행

```python
# 중복 값 삭제하기
revised_data["Age"].drop_duplicates()
```

## 데이터 그룹화 객체 만들기

groupby() 매서드 실행

```python
# 연령대 데이터 그룹화하기
revised_data.groupby(["Age"])
```

## 연령대 별 응답 숫자 확인하기

groupby.size() 메서드 실행  
ex) data.groupby(["Age"]).size()

```python
# 연령대 그룹별 크기 확인하기
size_by_age = revised_data.groupby(["Age"]).size()
size_by_age
```

## 국가별 응답 숫자 확인하기

```python
# 국가 그룹별 크기 확인하기
size_by_country = revised_data.groupby(["Country"]).size()
size_by_country
```

## 선그래프

plot() / plot.line() 메서드 실행

```python
# 선 그래프 그리기
size_by_age.plot.line(rot=45)
```

## 수직 막대그래프
plot.bar() 

```python
# 수직 막대그래프 그리기
size_by_age.plot.bar()
```

# 많이 사용되는 프로그래밍 언어 그래프 그리기
## 특정 열 추출하기

```python
# 프로그래밍 언어 데이터 추출
languages = revised_data["LanguageHaveWorkedWith"]
languages
```

## 단일 객체를 리스트로 변한하기
split() 메서드 실행

```python
# 데이터 문자열 변환 후 구분자(;)로 구분
languages = languages.str.split(";")
languages
```

## 리스트의 각 항목을 열로 나누기
explode() 메서드 실행

```python
# 리스트 항목을 행으로 나누기
exploded_languages = languages.explode()
exploded_languages
```

## 프로그래밍별 응답 수 확인하기

```python
# 프로그래밍별 응답 수 구하기
size_by_languages = exploded_languages.groupby(exploded_languages).size()
size_by_languages
```

```python
size_by_languages.nlargest(10).plot.pie(figsize=(10, 10))
```

## 백분율 표기하기
autpoct() 메서드 실행  
plot.pie(figsize=(10, 10), autopct = '%1.0f%%')

```python
size_by_languages.nlargest(10).plot.pie(figsize=(10, 10), autopct='%1.0f%%')
```
